### Modificando classes de variáveis

```r
# Nascimento: É possivel criar novas variáveis a partir desta, tais como ano de nascimento, mês e dia.

df$ano_nascimento = lubridate::year(df$data_nascimento)
df$mes_nascimento = lubridate::month(df$data_nascimento)
df$dia_nascimento = lubridate::day(df$data_nascimento)

df$data_nascimento = NULL
```

```r
# Fumante
df$fumante %>% unique()
df$fumante = factor(df$fumante)
```

```r
# Signo
df$signo %>% unique()
df$signo = factor(df$signo,
                  levels = c("Áries","Touro","Gêmeos","Câncer","Leão","Virgem",
                             "Libra","Escorpião","Sagitário","Capricórnio",
                             "Aquário","Peixes"))
```

```r
# Região
df$regiao %>% unique()
df$regiao = factor(df$regiao, levels = seq(4))
```

```r
# Facebook
df$facebook %>% unique()
df$facebook = factor(df$facebook)
```

```r
# classe
df$classe %>% unique() %>% sort()
df$classe = factor(df$classe, levels = seq(10))
```

```r
# Exclui-se a matricula da base, pois é uma variável chave do banco de dados que identifica o cliente.
df$id_matricula = NULL
```

# Análise das variáveis categóricas

Para analisar as influências das variáveis categóricas com o custo, foi utilizado de testes de hipóteses.

Premissas: 
* Como o custo não segue uma distribuição normal, é aconselhado utilizar métodos não paramétricos para validação.
* Para variáveis categóricas com dois grupos utizou-se Teste de Mann-Whitney.
* Para variáveis com mais de dois grupos utilizou-se Teste de Kruskal-Wallis.

```r
# Para a variável Fumante o teste rejeitou a hipótese de igualdade entre medianas,
# sendo assim a variável fumante possui influência no custo.
# Obs.: Não Fumante = 0, Fumante = 1
options(repr.plot.width=10, repr.plot.height=4)
wilcox.test(custo ~ fumante, df) 

p1 = ggplot(data = na.omit(df), aes(x = fumante,y = custo)) + 
  geom_boxplot(fill = "#4286f5") + 
  labs(title = "Gráfico de Boxplot do Custo por Fumante", x = "Fumante", y = "Custo")
p2 = ggplot(na.omit(df), aes(x = custo)) + 
  geom_histogram(aes(y = ..density..), fill="#4286f5", bins = 30) +
  geom_density(alpha=0, color ="#ff000b") + 
  labs(title = "Gráfico de distribuição do Custo por Fumante", x = "Custo", y = "Densidade") + 
facet_wrap(~fumante)

grid.arrange(
    p1,
    p2,
    ncol = 2
  )
```

```r
# Para a variável Sexo o teste aceitou a hipótese de igualdade entre medianas,
# sendo assim a variável sexo não possui influência no custo.
# Obs.: M = 0, F = 1
options(repr.plot.width=10, repr.plot.height=4)
wilcox.test(custo ~ sexo, df) 

p1 = ggplot(data = na.omit(df), aes(x = sexo,y = custo)) + 
  geom_boxplot(fill = "#4286f5") + 
  labs(title = "Gráfico de Boxplot do Custo por Sexo", x = "Sexo", y = "Custo")
p2 = ggplot(na.omit(df), aes(x = custo)) + 
  geom_histogram(aes(y = ..density..), fill="#4286f5", bins = 30) +
  geom_density(alpha=0, color ="#ff000b") + 
  labs(title = "Gráfico de distribuição do Custo por Sexo", x = "Custo", y = "Densidade") + 
facet_wrap(~sexo)

grid.arrange(
    p1,
    p2,
    ncol = 2
  )
```

# Data Imputation / Cleaning

```r
# Para evitar que percamos uma parte do dataset original excluindo todas as matrículas que possuem valores nulos
# foi aplicado o data imputation somente nas variáveis selecionadas para o modelo.

# Variáveis selecionadas: fumante,idade e imc

# Como temos a data de nascimento podemos prever a idade com a data de nascimento.

summary(lm(idade ~ ano_nascimento + mes_nascimento + dia_nascimento, df))

df$idade = ifelse(!is.na(df$idade),df$idade,
                  (2020.36398 -1.00013*df$ano_nascimento -0.11294*df$mes_nascimento -0.00149*df$dia_nascimento))

# A variável IMC foi imputada a partir da média de IMC por fumante e idade.

df$imc <- ave(df$imc,paste0(df$fumante,"-",df$idade),FUN=function(x) 
  ifelse(is.na(x), mean(x,na.rm=TRUE), x))

# Por ser uma variável sensível e o percentual de NA ser muito baixo para fumante,
# escolhi por não preencher valores de fumantes vazios e elimidar estes dados da base.

# A variável fumante foi transformada em variável númerica porque alguns métodos nâo aceitam variáveis categóricas.
              
df$fumante = as.numeric(as.character(df$fumante))
          
# Os valores 0 em custo foram retirados da base, pois possuem somente 3 matrículas. 
# Para saber se é um valor real seria necessário validar estes dados com a área de negócio.
              
df_final = df %>% 
  dplyr::select(custo, fumante, idade, imc) %>% 
  dplyr::filter(custo > 0) %>% 
  na.omit()
       
# Percentual de valores faltantes no dataset final em relação ao dataset original
              
print(paste0("Percentual de dados excluíndo valores nulos de todas as variáveis da base: ",round((1-NROW(na.omit(df))/NROW(df)) * 100,2),"%"))
print(paste0("Percentual de dados excluíndo valores nulos somente das variáveis selecionadas: ",round((1-(NROW(df_final)/NROW(df))) * 100,2),"%"))
```

## Treinamento dos modelos

```r
# Divisão da base de treino e teste
set.seed(123)
divisao = sample.split(df_final$custo, SplitRatio = 0.7)
df_train = subset(df_final,divisao)
df_test = subset(df_final,!divisao)
```

```r
# Função para análise dos resultados dos modelos

result_models <- function(y_pred_train,y_pred_test, y_true_train,y_true_test,modelo){
    
  options(repr.plot.width=10, repr.plot.height=6)
  
  result_table_train <- data.frame(
    RMSE = round(MLmetrics::RMSE(y_pred = y_pred_train,y_true = y_true_train),2),
    MAE = round(MLmetrics::MAE(y_pred = y_pred_train,y_true = y_true_train),2),
    MAPE = round(MLmetrics::MAPE(y_pred = y_pred_train,y_true = y_true_train)*100,2),
    R2 = round(MLmetrics::R2_Score(y_pred = y_pred_train,y_true = y_true_train),2),
    p_value_residuos  = round(shapiro.test(y_true_train - y_pred_train)$p.value,3)
  )
  rownames(result_table_train) <- "Train"
  
  result_table_test <- data.frame(
    RMSE = round(MLmetrics::RMSE(y_pred = y_pred_test,y_true = y_true_test),2),
    MAE = round(MLmetrics::MAE(y_pred = y_pred_test,y_true = y_true_test),2),
    MAPE = round(MLmetrics::MAPE(y_pred = y_pred_test,y_true = y_true_test)*100,2),
    R2 = round(MLmetrics::R2_Score(y_pred = y_pred_test,y_true = y_true_test),2),
    p_value_residuos = round(shapiro.test(y_true_test - y_pred_test)$p.value,3)
  )
  rownames(result_table_test) <- "Test"
  
  train_plot <- ggplot() + geom_point(aes(x = y_pred_train,y = y_true_train), color = "#4286f5") +
    geom_smooth(aes(x = y_pred_train,y = y_true_train), color = "#ff000b", method = "lm") + 
    labs(title = "Base Treino - Custo Real vs. Custo Predito", x = "Custo Predito", y = "Custo Real")
  
  test_plot <- ggplot() + geom_point(aes(x = y_pred_test,y = y_true_test), color = "#4286f5") +
    geom_smooth(aes(x = y_pred_test,y = y_true_test), color = "#ff000b", method = "lm") + 
    labs(title = "Base Teste - Custo Real vs. Custo Predito", x = "Custo Predito", y = "Custo Real")
  
  
  residual_plot_test <- ggplot() + 
    geom_point(aes(x = y_pred_test,y = y_true_test - y_pred_test), color = "#4286f5") + 
    geom_hline(yintercept = 0, color = "#ff000b")+
    labs(title = "Base Teste - Resíduos vs. Custo Predito", x = "Custo Predito", y = "Resíduos")
  
  
  normal_residual_plot_test <- ggplot() + 
    geom_qq(aes(sample=y_true_test - y_pred_test), color = "#4286f5") + 
    geom_qq_line(aes(sample=y_true_test - y_pred_test), color = "#ff000b") +
    labs(title = "Base Teste - Normal QQ", x = "Teórico", y = "Resíduos")
  
  
  
  grid.arrange(
    tableGrob(result_table_train),
    tableGrob(result_table_test),
    train_plot,
    test_plot,
    residual_plot_test,
    normal_residual_plot_test,
    ncol = 2,
    nrow = 3,
    heights = c(0.5,1,1),
    clip = FALSE
  )
  
  test_results = result_table_test
  test_results$modelo = modelo
  
  return(test_results)
  
}
```

### Modelo Regressão Linear

```r
df_train_lm = df_train
df_train_lm$idade_fumante = df_train_lm$idade * df_train_lm$fumante
df_train_lm$imc_fumante = df_train_lm$imc * df_train_lm$fumante

df_test_lm = df_test
df_test_lm$idade_fumante = df_test_lm$idade * df_test_lm$fumante
df_test_lm$imc_fumante = df_test_lm$imc * df_test_lm$fumante

lm_model = lm(custo ~ . , df_train_lm)
summary(lm_model)
lm_predict_train = predict(lm_model,df_train_lm %>% select(-custo))
lm_predict_test = predict(lm_model,df_test_lm %>% select(-custo))

lm_results = result_models(y_pred_train = lm_predict_train ,y_pred_test = lm_predict_test,
              y_true_train = df_train$custo,y_true_test = df_test$custo,
              modelo = "Linear Regression")
```

Obs.: O modelo regressão linear foi escolhido como modelo referência devido a sua alta interpretabilidade e de natureza analítica.

### Modelo Decision Tree

```r
# Ajuste de hiperparâmetros
set.seed(123)
dt_grid <- tune.rpart(formula = custo ~ .,
                           data = df_train,
                           maxdepth = c(1,2,3,4,5),
                           cp = c(0.002,0.005,0.01,0.015,0.02,0.03)
)
summary(dt_grid)
dt_grid$best.parameters
```

```r
dt_model = rpart(custo ~ ., df_train, maxdepth = 4, cp = 0.002)

dt_predict_train = predict(dt_model,df_train %>% select(-custo))
dt_predict_test = predict(dt_model,df_test %>% select(-custo))

dt_results = result_models(y_pred_train = dt_predict_train ,y_pred_test = dt_predict_test,
              y_true_train = df_train$custo,y_true_test = df_test$custo,
              modelo = "Decision Tree")
```

### Modelo KNN

```r
# Ajuste de hiperparâmetros

df_train_scaled = df_train
df_train_scaled = scale(df_train_scaled) %>% as.data.frame()
df_train_scaled$custo = df_train$custo


df_test_scaled = df_test
df_test_scaled = scale(df_test_scaled) %>% as.data.frame()
df_test_scaled$custo = df_test$custo


grid_knn_params  = expand.grid(method = c("Euclidean", "Manhattan"),
                               k = c(2,3,4,5,6,7,8,9,10))

grid_knn_results = sapply(seq(NROW(grid_knn_params)), function(i){
  
  modelo =  gknn(x = df_train_scaled %>% select(-custo),
                    y = df_train$custo,
                    k = grid_knn_params$k[i],
                    method = grid_knn_params$method[i])
  prediction = predict(modelo,df_train_scaled %>% select(-custo))
  MLmetrics::RMSE(y_pred = prediction,y_true = df_train$custo)
  
})

knn_best_params <- grid_knn_params[which.min(grid_knn_results),]
knn_best_params
```

```r
knn_model =  caret::knnreg(x = df_train_scaled %>% select(-custo),
                  y = df_train$custo,
                  k = 8,
                  method = "Euclidean")


knn_predict_train = predict(knn_model,df_train_scaled %>% select(-custo))
knn_predict_test = predict(knn_model,df_test_scaled %>% select(-custo))

knn_results = result_models(y_pred_train = knn_predict_train ,y_pred_test = knn_predict_test,
              y_true_train = df_train$custo,y_true_test = df_test$custo,
              modelo = "KNN")
```

### Modelo xgboost

```r
# Ajuste de hiperparâmetros
set.seed(123)
grid_xgb_params  = expand.grid(nrounds = c(30,50),
                               eta = c(0.01, 0.1),
                               gamma = c(0,3,5),
                               subsample = c(0.6,0.7,0.8),
                               max_depth = c(3,4,5)
                               )

grid_xgb_results = sapply(seq(NROW(grid_xgb_params)), function(i){
  
  modelo = xgboost(data = data.matrix(df_train %>% select(-custo)),
                   label = df_train$custo,
                   objective = "reg:squarederror",
                   nrounds = grid_xgb_params$nrounds[i],
                   eta = grid_xgb_params$eta[i],
                   subsample = grid_xgb_params$subsample[i],
                   max_depth = grid_xgb_params$max_depth[i],
                   verbose = 0
  )
    
  prediction = predict(modelo,data.matrix(df_train %>% select(-custo)))
  MLmetrics::RMSE(y_pred = prediction,y_true = df_train$custo)
  
})

xgb_best_params <- grid_xgb_params[which.min(grid_xgb_results),]
xgb_best_params
```

```r
set.seed(123)
xgb_model = xgboost(data = data.matrix(df_train %>% select(-custo)),
                    label = df_train$custo,
                    objective = "reg:squarederror",
                    nrounds = 50,
                    gamma = 0,
                    eta = 0.1,
                    subsample = 0.8,
                    max_depth = 5,
                    eval_metric = c("rmse","mae")
)


xgb_predict_train = predict(xgb_model,data.matrix(df_train %>% select(-custo)))
xgb_predict_test = predict(xgb_model,data.matrix(df_test %>% select(-custo)))

xgb_results = result_models(y_pred_train = xgb_predict_train ,y_pred_test = xgb_predict_test,
                            y_true_train = df_train$custo,y_true_test = df_test$custo,
                            modelo = "XGBoost")
```

Em todos os modelos não foi possível ajustar os resíduos para uma distribuição normal.

```r
# Aplicando Data Imputation conforme feito na base de treino do modelo
df_teste_final$ano_nascimento = lubridate::year(df_teste_final$data_nascimento)
df_teste_final$mes_nascimento = lubridate::month(df_teste_final$data_nascimento)
df_teste_final$dia_nascimento = lubridate::day(df_teste_final$data_nascimento)

# Calculando Idade para valores nulos
df_teste_final$idade = ifelse(!is.na(df_teste_final$idade),df_teste_final$idade,
(2020.36398 -1.00013*df_teste_final$ano_nascimento -0.11294*df_teste_final$mes_nascimento -0.00149*df_teste_final$dia_nascimento))

# Calculando IMC a partir da média de IMC por fumante e idade.

df_teste_final$imc <- ave(df_teste_final$imc,paste0(df_teste_final$fumante,"-",df_teste_final$idade),FUN=function(x) 
  ifelse(is.na(x), mean(x,na.rm=TRUE), x))

# Verificando valores nulos na variáveis preditoras
colSums(is.na(df_teste_final %>% select(imc,idade,fumante)))
```
