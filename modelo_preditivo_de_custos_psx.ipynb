# Análise das variáveis categóricas

Para analisar as influências das variáveis categóricas com o custo, foi utilizado de testes de hipóteses.

Premissas: 
* Como o custo não segue uma distribuição normal, é aconselhado utilizar métodos não paramétricos para validação.
* Para variáveis categóricas com dois grupos utizou-se Teste de Mann-Whitney.
* Para variáveis com mais de dois grupos utilizou-se Teste de Kruskal-Wallis.

```r
# Para a variável Fumante o teste rejeitou a hipótese de igualdade entre medianas,
# sendo assim a variável fumante possui influência no custo.
# Obs.: Não Fumante = 0, Fumante = 1
options(repr.plot.width=10, repr.plot.height=4)
wilcox.test(log_custo ~ fumante, df_eda) 

p1 = ggplot(data = na.omit(df_eda), aes(x = fumante,y = log_custo)) + 
  geom_boxplot(fill = "#4286f5") + 
  labs(title = "Gráfico de Boxplot do log10(Custo) por Fumante", x = "Fumante", y = "log10(Custo)")
p2 = ggplot(na.omit(df_eda), aes(x = log_custo)) + 
  geom_histogram(aes(y = ..density..), fill="#4286f5", bins = 30) +
  geom_density(alpha=0, color ="#ff000b") + 
  labs(title = "Gráfico de distribuição do log10(Custo) por Fumante", x = "log10(Custo)", y = "Densidade") + 
facet_wrap(~fumante)

grid.arrange(
    p1,
    p2,
    ncol = 2
  )
```

```r
# Para a variável Sexo o teste aceitou a hipótese de igualdade entre medianas,
# sendo assim a variável sexo não possui influência no custo.
# Obs.: M = 0, F = 1
options(repr.plot.width=10, repr.plot.height=4)
wilcox.test(log_custo ~ sexo, df_eda) 

p1 = ggplot(data = na.omit(df_eda), aes(x = sexo,y = log_custo)) + 
  geom_boxplot(fill = "#4286f5") + 
  labs(title = "Gráfico de Boxplot do log10(Custo) por Sexo", x = "Sexo", y = "log10(Custo)")
p2 = ggplot(na.omit(df_eda), aes(x = log_custo)) + 
  geom_histogram(aes(y = ..density..), fill="#4286f5", bins = 30) +
  geom_density(alpha=0, color ="#ff000b") + 
  labs(title = "Gráfico de distribuição do log10(Custo) por Sexo", x = "log10(Custo)", y = "Densidade") + 
facet_wrap(~sexo)

grid.arrange(
    p1,
    p2,
    ncol = 2
  )
```

# Data Imputation / Cleaning

```r
# Para evitar que percamos uma parte do dataset original excluindo todas as matrículas que possuem valores nulos
# foi aplicado o data imputation somente nas variáveis selecionadas para o modelo.

# Variáveis selecionadas: fumante,idade e imc

# Como temos a data de nascimento podemos prever a idade com a data de nascimento.

summary(lm(idade ~ ano_nascimento + mes_nascimento + dia_nascimento, df))

df_eda$idade = ifelse(!is.na(df_eda$idade),df_eda$idade,
                  (2020.36398 -1.00013*df_eda$ano_nascimento -0.11294*df_eda$mes_nascimento -0.00149*df_eda$dia_nascimento))

# A variável IMC foi imputada a partir da média de IMC por fumante e idade.

df_eda$imc <- ave(df_eda$imc,paste0(df_eda$fumante,"-",df_eda$idade),FUN=function(x) 
  ifelse(is.na(x), mean(x,na.rm=TRUE), x))

# Por ser uma variável sensível e o percentual de NA ser muito baixo para fumante e filhos,
# escolhi por não realizar o data Imputation nestas variáveis e elimidar estes dados da base.


# A variável fumante foi transformada em variável númerica porque alguns métodos nâo aceitam variáveis categóricas.
              
df_eda$fumante = as.numeric(as.character(df_eda$fumante))
              
df_final = df_eda %>% 
  dplyr::select(custo, fumante, idade, imc, filhos) %>% 
  na.omit()
       
# Percentual de valores faltantes no dataset final em relação ao dataset original
              
print(paste0("Percentual de dados excluíndo valores nulos de todas as variáveis da base: ",round((1-NROW(na.omit(df))/NROW(df)) * 100,2),"%"))
print(paste0("Percentual de dados excluíndo valores nulos somente das variáveis selecionadas: ",round((1-(NROW(df_final)/NROW(df))) * 100,2),"%"))
```

## Treinamento dos modelos

```r
# Divisão da base de treino e teste
set.seed(123)
divisao = sample.split(df_final$custo, SplitRatio = 0.7)
df_train = subset(df_final,divisao)
df_test = subset(df_final,!divisao)

real_train = df_train$custo
real_test = df_test$custo

df_train$custo = log10(df_train$custo)
df_test$custo = log10(df_test$custo)
```

```r
# Função para análise dos resultados dos modelos

result_models <- function(y_pred_train,y_pred_test, y_true_train,y_true_test,modelo){
    
  options(repr.plot.width=10, repr.plot.height=6)
  
  result_table_train <- data.frame(
    RMSE = round(MLmetrics::RMSE(y_pred = y_pred_train,y_true = y_true_train),2),
    MAE = round(MLmetrics::MAE(y_pred = y_pred_train,y_true = y_true_train),2),
    MAPE = round(MLmetrics::MAPE(y_pred = y_pred_train,y_true = y_true_train)*100,2),
    R2 = round(MLmetrics::R2_Score(y_pred = y_pred_train,y_true = y_true_train),2),
    sd_res = round(sd(sqrt((y_true_train-y_pred_train)^2)),2),
    p_value_res  = round(shapiro.test(y_true_train - y_pred_train)$p.value,5)
  )
  rownames(result_table_train) <- "Train"
  
  result_table_test <- data.frame(
    RMSE = round(MLmetrics::RMSE(y_pred = y_pred_test,y_true = y_true_test),2),
    MAE = round(MLmetrics::MAE(y_pred = y_pred_test,y_true = y_true_test),2),
    MAPE = round(MLmetrics::MAPE(y_pred = y_pred_test,y_true = y_true_test)*100,2),
    R2 = round(MLmetrics::R2_Score(y_pred = y_pred_test,y_true = y_true_test),2),
    sd_res = round(sd(sqrt((y_true_test-y_pred_test)^2)),2),
    p_value_res = round(shapiro.test(y_true_test - y_pred_test)$p.value,5)
  )
  rownames(result_table_test) <- "Test"
  
  train_plot <- ggplot() + geom_point(aes(x = y_pred_train,y = y_true_train), color = "#4286f5") +
    geom_smooth(aes(x = y_pred_train,y = y_true_train), color = "#ff000b", method = "lm") + 
    labs(title = "Base Treino - Custo Real vs. Custo Predito", x = "Custo Predito", y = "Custo Real")
  
  test_plot <- ggplot() + geom_point(aes(x = y_pred_test,y = y_true_test), color = "#4286f5") +
    geom_smooth(aes(x = y_pred_test,y = y_true_test), color = "#ff000b", method = "lm") + 
    labs(title = "Base Teste - Custo Real vs. Custo Predito", x = "Custo Predito", y = "Custo Real")
  
  
  residual_plot_test <- ggplot() + 
    geom_point(aes(x = y_pred_test,y = y_true_test - y_pred_test), color = "#4286f5") + 
    geom_hline(yintercept = 0, color = "#ff000b")+
    labs(title = "Base Teste - Resíduos vs. Custo Predito", x = "Custo Predito", y = "Resíduos")
  
  
  normal_residual_plot_test <- ggplot() + 
    geom_qq(aes(sample=y_true_test - y_pred_test), color = "#4286f5") + 
    geom_qq_line(aes(sample=y_true_test - y_pred_test), color = "#ff000b") +
    labs(title = "Base Teste - Normal QQ", x = "Teórico", y = "Resíduos")
  
  
  
  grid.arrange(
    tableGrob(result_table_train),
    tableGrob(result_table_test),
    train_plot,
    test_plot,
    residual_plot_test,
    normal_residual_plot_test,
    ncol = 2,
    nrow = 3,
    heights = c(0.5,1,1),
    clip = FALSE
  )
  
  test_results = result_table_test
  test_results$modelo = modelo
  
  return(test_results)
  
}
```

### Modelo Regressão Linear

```r
df_train_lm = df_train
df_train_lm$idade_fumante = df_train_lm$idade * df_train_lm$fumante
df_train_lm$imc_fumante = df_train_lm$imc * df_train_lm$fumante
df_train_lm$filhos_fumante = df_train_lm$filhos * df_train_lm$fumante

df_test_lm = df_test
df_test_lm$idade_fumante = df_test_lm$idade * df_test_lm$fumante
df_test_lm$imc_fumante = df_test_lm$imc * df_test_lm$fumante
df_test_lm$filhos_fumante = df_test_lm$filhos * df_test_lm$fumante

lm_model = lm(custo ~ . - imc , df_train_lm)
summary(lm_model)
lm_predict_train = 10^(predict(lm_model,df_train_lm %>% select(-custo)))
lm_predict_test = 10^(predict(lm_model,df_test_lm %>% select(-custo)))

lm_results = result_models(y_pred_train = lm_predict_train ,y_pred_test = lm_predict_test,
              y_true_train = real_train,y_true_test = real_test,
              modelo = "Linear Regression")
```

Obs.: O modelo regressão linear foi escolhido como modelo referência devido a sua alta interpretabilidade e de natureza analítica.
